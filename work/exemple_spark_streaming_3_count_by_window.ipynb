```python
# imports
import socket
from pyspark.sql import SparkSession
from pyspark.streaming import StreamingContext

def get_time():
    from datetime import datetime
    return "%s min %s s"%(datetime.now().minute,datetime.now().second)

def print_time():
    print(get_time())
# paramètrage
listen_to_ip   = socket.gethostname()
with open("port.param", "r") as f:
    port           = f.readline()
    listen_to_port = int(port)
print("Going to listen on %s:%s"%(listen_to_ip, listen_to_port))
nb_secondes    = 1
batchDuration  = nb_secondes
checkpoint_dir = "./checkpoint/"
print_time()
# Initialisation de spatrk
spark       = SparkSession.builder.getOrCreate()
sc          = spark.sparkContext
print_time()
# création du streaming context
ssc         = StreamingContext(sc, nb_secondes)
dstream     = ssc.socketTextStream(listen_to_ip, listen_to_port)
ssc.checkpoint(checkpoint_dir)
print_time()

# création du Directed Acyclic Graph :
data            = dstream.map(lambda x: x.split(","))
clients_facture = data.map(lambda x: (x[0], float(x[1])) )
slideDuration   = batchDuration 
windowDuration  = batchDuration 
update_client   = clients_facture.reduceByKeyAndWindow(
                        func           = lambda x,y : x+y   , 
                        invFunc        = lambda x,y : x-y   , 
                        windowDuration = windowDuration * 3 , 
                        slideDuration  = slideDuration      )
update_client.pprint()

message = """
On va calculer chaque {slideDuration} seconde la somme des valeurs affectées 
par entreprise pendant les dernières {windowDuration} secondes
passées""".format(**locals())
# démarrage de l'écoute de la socket : 
print(message)
print_time()
ssc.start()
```
